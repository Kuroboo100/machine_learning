```python
import numpy as np
import matplotlib.pyplot as plt
import inspect
%matplotlib inline
```

```python
x=np.arange(-10,10,0.1)
y=np.cos(x)
z=np.sin(x)
```

```python
fig = plt.figure(figsize=(13,1))
#pltfigurefig 
ax1 = fig.add_subplot(221) # 211
#figureadd?subplotax1
ax1.plot(x,y,label="cos")
ax1.plot(x,z,label="sin")
ax1.set_title("cos &sin")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.legend()

ax2 = fig.add_subplot(222) # 212 
#figureadd?subplotax2
ax2.plot(x,y*z)
ax2.set_title("cos *sin")
ax2.set_xlabel("x")
ax2.set_ylabel("y")

ax3 = fig.add_subplot(223) 
ax3.plot(x,y*z**2)
ax3.set_title("cos *sin^2")
ax3.set_xlabel("x")
ax3.set_ylabel("y")

ax4 = fig.add_subplot(224)
ax4.plot(x,y*z**3)
ax4.set_title("cos *sin^3")
ax4.set_xlabel("x")
ax4.set_ylabel("y")

plt.subplots_adjust(top=10)
```

```python
fig4,axes=plt.subplots(1,2,figsize=(10,3))

#
i=np.arange(1,10,0.1)
f=lambda i:np.sin(i)
g=lambda i:np.cos(i)
func=[f,g]
func_name=("sin","cos")

fig4.suptitle("sin cos")
for ax,model,name in zip(axes,func,func_name):
    ax.plot(i,model(i))
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_title(str(name))
```
