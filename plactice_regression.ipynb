```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection  import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
import mglearn
```

```python
X,y=mglearn.datasets.make_wave()
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
coef=lr.coef_
intercept=lr.intercept_
```

```python
x_=np.arange(-10,10,0.1)
y_=coef[0]*x_+intercept
```

```python
fig,ax=plt.subplots(1,1)
ax.scatter(X,y)
ax.plot(x_,y_)
ax.set_xlim(-3,3)
ax.set_ylim(-3,3)
ax.set_xlabel("feature")
ax.set_ylabel("target")
ax.set_title("train accuracy ="+'{:.3f}'.format(lr.score(X_train,y_train))+"  test accuracy ="+'{:.3f}'.format(lr.score(X_test,y_test)))
plt.show()
```

```python
rg=Ridge(alpha=0.1).fit(X_train,y_train)
ac_train=rg.score(X_train,y_train)
ac_test=rg.score(X_test,y_test)
```
